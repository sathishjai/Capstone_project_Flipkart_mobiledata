```python
# Impute missing values
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(df[['Ratings_Stars','Ram','Rom','Camera_count','Inch','Price']])

# Convert imputed data back to DataFrame
imputed_df = pd.DataFrame(imputed_data, columns=['Ratings_Stars','Ram','Rom','Camera_count','Inch','Price'])

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3)  # Define the number of clusters
kmeans.fit(imputed_df[['Ratings_Stars','Ram','Rom','Camera_count','Inch','Price']])  # Fit the model

# Predict clusters for data points
clusters = kmeans.predict(imputed_df[['Ratings_Stars','Ram','Rom','Camera_count','Inch','Price']])

# Access cluster centers
cluster_centers = kmeans.cluster_centers_
print("Cluster Centers:")
print(cluster_centers)

# Plot the data with cluster assignments
plt.scatter(imputed_df['Inch'], imputed_df['Price'], c=clusters, cmap='viridis')

# Add cluster centers to the plot (assuming 'Price' is last feature)
plt.scatter([0]*len(cluster_centers), cluster_centers[:, -1], marker='*', s=200, c='red', label='Cluster Centers')
plt.xlabel('Inch')
plt.ylabel('Price')  # Replace with actual feature name if different
plt.title('Clustered Data')
plt.legend()
plt.show()
```
